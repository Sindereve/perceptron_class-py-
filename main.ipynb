```python
#import
from perceptron import Perceptron
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os 
```

Input data set
 
We use standart data of 'iris'

```python
df = pd.read_csv('iris.data', header=None, encoding='utf-8')
print(df.shape)
df.head()
```

analyse data

```python
#change data in y and X
y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)
X = df.iloc[0:100, [0,2]].values

plt.scatter(X[:50,0],X[:50,1],
    color='red', marker = 'x', label='bristly')

plt.scatter(X[50:100,0],X[50:100,1],
    color='blue', marker = 'o', label='multicoloured')

plt.xlabel('size sepals')
plt.ylabel('size petal')
plt.legend(loc='upper left')
plt.show()
```

```python
ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(X,y)

plt.plot(range(1, len(ppn.errors_)+1),ppn.errors_, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Number of errors')
plt.show()
```

Checking the operation of the prediction model

```python
from sklearn import metrics

y_predict = ppn.predict(X)
print(f'Result real predict {metrics.r2_score(y,y_predict)}')
```
